```python
# RETO 1: Practicar la creación de Series y DataFrames desde cero, y aplicar el flujo
# de trabajo de Carga e Inspección en un dataset nuevo.
import numpy as np
import pandas as pd

#Serie de Pandas llamada s_calificaciones
calificaciones = {'Ana': 10, 'Luis': 7.5, 'Marta': 9.0, 'Juan': 5.5}
s_calificaciones = pd.Series(calificaciones)

#Serie nueva sumando 0.5 a la serie original
s_calificaciones_final = s_calificaciones + 0.5

#Imprimiendo la serie
print(s_calificaciones_final)
```

```python
#DataFrame llamado df_inventario
data = {
    'producto': ['Laptop', 'Mouse', 'Teclado', 'Monitor'],
    'stock': [20, 150, 80, 35],
    'precio_usd': [1200, 25, 60, 300]
}

df_inventario = pd.DataFrame(data)

#Imprimiendo el DataFrame
print("\n",df_inventario)
```

```python
#Uso del método .set_index() para hacer que la columna producto sea el nuevo índice de filas
df_inventario.set_index('producto', inplace=True)

#Imprimiendo el DataFrame modificado
print(df_inventario)
```

```python
#Usando un dataset de Ventas de Videojuegos
import pandas as pd
# Cargar los datos mediante URL
try:
    url = 'https://gist.githubusercontent.com/designernatan/27da044c6dc823f7ac7fe3a01f4513ed/raw/d15b5c7d7a5efb38750b16ec935fc126ec9a6e79/vgsales.csv'
    df_juegos = pd.read_csv(url)
    print("DataFrame 'df_juegos' CREADO exitosamente.")
except Exception as e:
    print(f"¡ERROR AL CREAR! El problema es: {e}")

# Paso 1: Imprimir las primeras 10 filas
try:
    print("\n--- Imprimiendo las primeras 10 filas ---")
    print(df_juegos.head(10))
    print("\n¡Se ha realizado con éxito!.")
except NameError:
    print("\nEl 'NameError' persiste. Esto significa que el Paso 2 falló.")
except Exception as e:
    print(f"¡ERROR AL CREAR! El problema es: {e}")

#Paso 2: Imprimir la forma (.shape)
print("\n--- Uso de .shape ---")
print(f"Forma del DataFrame: {df_juegos.shape}")
print(f"El dataset contiene {df_juegos.shape[0]} filas")

# Paso 3: Ejecutar .info()
print("\n--- Uso de .info()")
df_juegos.info()
print(f"Las columnas 'Year' y 'Publisher' tienen valores nulos.")

#Paso 4: Ejecutar .describe()
print("\n--- Uso de .describe()")
df_juegos.describe()
print(f"El año más reciente en el archivo es el 2020")
print(f"El promedio de ventas en Norteamérica es de 0.264667")

#Paso 5: Ejecutar .columns
print("\n--- Imprimiendo la lista de nombres de columnas")
print(df_juegos.columns)
```
